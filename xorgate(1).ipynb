```
import numpy as np
import os 


# data = np.array([[0,0],[0,1],[1,0],[1,1]])
# n = 2

# m = 4

# y = np.zeros(m)

# #n = np.random.rand

# for j in range(y.size):
    
#     if data[j,0] != data[j,1]:
#         y[j] = 1
    
    
# #reshaping only for this eq.
# data = data.T
# #weights  w1 -(2 * 2 ) w2 - (1 * 2)
# #w1 = np.array([[-20,20],[20,-20]])
# #w2 = np.array([[20,20]])



# print(data.shape,y.shape)



x  = np.array([[0,0],[1,0],[0,1],[1,1]]).reshape(4,2)
y = np.array([0,1,1,0]).reshape(1,4)

print(x.shape,y.shape)


def sigmoid(a):
      return 1/(1 + np.exp(-a))

class XOR:

  def __init__(self):

    np.random.seed(0)
    self.w1 = np.random.rand(2,2)

    #print(w1)
    np.random.seed(1)
    self.w2 = np.random.rand(1,2)

    #print(w2)

    #b1 = np.array([[-10],[-10]])
    np.random.seed(2)
    self.b2 = np.random.rand()

    np.random.seed(3)
    self.b1 = np.random.rand(1,2)


    self.alpha = 0.03
    jloss = 0


  def fit(self,X,Y):

    for i in range(10000):
    #feedword
      
      # node 1
      h1 = np.dot(x,self.w1) + self.b1# ( 2,2) * (2,4) = (4,2)
      a = sigmoid(h1) # (4,2)
    
    
      #node 2
      h2 = np.dot(self.w2,a.T) + self.b2 # (1,2) * (2,4) =(1,4)
      y_pred = sigmoid(h2) # - (1,4)
      output = np.round(y_pred.T)
      # print("INPUT = {} : Y_PRED = {} ,{}".format(data,y_pred, np.round(y_pred)))
      
      
      #cost function    
      jloss = np.mean(-(y* np.log2(y_pred) + (1-y)*np.log2(1-y_pred)))

      print("cost = {}".format(jloss))
      

      dz2 =  (y_pred - y)  # (1,4)
      
      # (2,4)
      
      dz1 = np.dot(dz2.T,self.w2) *   (a  * (1-a)) # (4,1). (1,2) * (4,2) = (4,2)

      # print(dz1.shape)
    
      #backpropogation
      #weights of layer 3 (w2)
      dw2 = np.dot(dz2,a)  # (1,4) * (4,2) = (1,2)
      
      db2 = np.sum(dz2,1,keepdims=True)
      
      dw1 = np.dot(x.T,dz1) # (2,4) * (4,2) = (2,2) 

      db1 = np.sum(dz1,0,keepdims=True)

      
    
      self.w1 = self.w1 - alpha * dw1 
      
      
      self.w2 = self.w2 - alpha * dw2 
      
      self.b2 = self.b2 - alpha * db2
      
      
      self.b1 = self.b1 - alpha * db1 

      
  def predict(self,x): 
    z1 = np.dot(x,self.w1) + self.b1
    #print("z1:{}".format(z1))
    q1 = sigmoid(z1)
    #print("q1:{}".format(q1))
    z2 = np.dot(q1,self.w2.T) + self.b2
    #print("z2:{}".format(z2))
    q2 = sigmoid(z2)
    # print("q2:{}".format(q2))
    # print(q2)
    # print("input {} and output:{}".format(x,np.round(q2)))     

    return np.round(q2)


model = XOR()
model.fit(x,y)
model.predict(x)
```
